# Load the raw data

```python
import pandas as pd
pd.set_option('display.max_columns', None)

data = pd.read_csv('data/london-suicide.csv')
data
```

# Load the shapefile

```python
import geopandas as gpd
import matplotlib.pyplot as plt
shapefile_borough = gpd.read_file("statistical-gis-boundaries-london/ESRI/London_Borough_Excluding_MHW.shp")
shapefile_borough.plot(figsize=(20, 7))
```

# Combining the data with the shapefile

```python
shapefile_borough.index = shapefile_borough['NAME']
data.index = data['name']


df = shapefile_borough.join(data)
df
```

```python
df.plot(column = '2022', figsize=(20, 7), legend = True, cmap = 'Blues')
plt.title('London Suicide Counts in 2022')
plt.axis('off')
```

# Creating cutpoints

```python
quantiles = data.drop(columns=['code', 'name']).stack().quantile([0.25, 0.5, .75, 1])
print(quantiles)
```

```python
data.drop(columns=['code', 'name']).stack()
```

```python

fig, axes = plt.subplots(nrows=2, ncols=6, figsize=(30, 10)) 
axes = axes.flatten() 

for i, year in enumerate(range(2011, 2023)):
    ax = axes[i]
    df.plot(column=str(year), 
            cmap='Blues', 
            edgecolor = '0.8',
            legend=True, 
            scheme='user_defined', 
            classification_kwds={'bins': quantiles.round(2).to_list()},
            ax=ax)
    ax.set_title(f'London Suicide Counts in {year}')
    ax.axis('off')

plt.tight_layout()
plt.show()
```
